```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
```

```python
train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test.csv")
```

```python
train.info()
```

```python
train.drop('Cabin', axis = 1,  inplace = True), test.drop('Cabin', axis = 1, inplace = True)
```

```python
train.info(), test.info()
```

```python
train.dropna(how = 'all', inplace = True), test.dropna(how = 'all', inplace = True)
```

```python
train.info(), test.info()
```

```python
train['Age'].interpolate(inplace = True), test['Age'].interpolate(inplace = True)
```

```python
train.info()
```

```python
test.info()
```

```python
train['Embarked'].fillna(method = 'ffill', inplace = True)
```

```python
train.info()
```

```python
train_columns = train.columns
test_columns = test.columns
```

```python
for i in train_columns:
  print(f"{i}: {train.loc[:, i].isnull().sum()}")
```

```python
for i in test_columns:
  print(f"{i}: {test.loc[:, i].isnull().sum()}")
```

```python
train.info(), test.info()
```

```python
obi_stan = test.drop(['PassengerId', 'Name'], axis = 1)
```

```python
X = train.drop(['PassengerId', 'Survived', 'Name'], axis = 1)
y = train['Survived']
```

```python
print(X.columns)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
X_train.info()
```

```python
X_val.info()
```

```python
ct = make_column_transformer(
    (MinMaxScaler(), [0, 2, 3, 4, 6]),
    (OneHotEncoder(handle_unknown = 'ignore'), [1, 5, 7])
)
```

```python
ct.fit(X_train)
```

```python
X_train_norm = ct.transform(X_train)
```

```python
X_val_norm = ct.transform(X_val)
```

```python
n_n = np.arange(1, 21)
n_n
```

```python
from sklearn.neighbors import KNeighborsClassifier
a = np.arange(1, 21)
v_a = np.arange(1, 21)
c = 0
for i in n_n:
  model = KNeighborsClassifier(n_neighbors = i)
  model.fit(X_train_norm, y_train)
  a[c] = model.score(X_train_norm, y_train)*100
  v_a[c] = model.score(X_val_norm, y_val)*100
  c +=1
```

```python
print(a)
print(v_a)
```

```python
barrier = plt.bar((np.arange(1,21)), a, tick_label = (np.arange(1, 21)), label = "Training Accuracy")
dumfer = plt.bar((np.arange(1,21)), v_a, tick_label = (np.arange(1, 21)), label = "Validation Accuracy", alpha = 0.7)
plt.legend()
plt.bar_label(barrier)
plt.bar_label(dumfer)
```

```python
from sklearn.ensemble import RandomForestClassifier
n_t = np.arange(1, 101)
a = []
v_a = []
c = 0
for i in range(1, len(n_t)):
  model = RandomForestClassifier(n_estimators = i)
  model.fit(X_train_norm, y_train)
  a.append(int(model.score(X_train_norm, y_train)*100))
  v_a.append(int(model.score(X_val_norm, y_val)*100))
  c = c+1
```

```python
plt.figure(figsize = (35, 10))
barrier = plt.bar((np.arange(1, 100)), a, tick_label = (np.arange(1, 100)), label = "Training Accuracy")
dumfer = plt.bar((np.arange(1, 100)), v_a, tick_label = (np.arange(1, 100)), label = "Validation Accuracy", alpha = 0.9)
plt.legend()
plt.bar_label(barrier)
plt.bar_label(dumfer)
```
